# Using ___ to make REMs with Python 

```python
from rasterio.merge import merge
from rasterio.plot import show
import rasterio as rio
from pathlib import Path
from utility_function import mosaic_list_of_rasters
from riverrem.REMMaker import REMMaker
```

## REM of boston


```python
# Get list of rasters
boston_path = Path('data/boston')
raster_files = list(boston_path.iterdir())
```

```python
boston_mosaic = "output/boston_mosaic.tiff"

if Path(boston_mosaic).is_file():
    print(f"File {boston_mosaic} exists already!")
else:
    print("Creating mosaic..")
    mosaic_list_of_rasters(raster_files, boston_mosaic )
```

```python
rem_maker = REMMaker(
    dem=boston_mosaic,
    out_dir="output"
                    )
rem_maker.make_rem()

rem_maker.make_rem_viz(cmap="mako_r")
```
